```python
# Import Packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Class_labels'] # As per the iris dataset information
```

```python
# Load the data
df = pd.read_csv('IRIS.csv', names=columns)
```

```python
df.head()
```

```python
# Some basic statistical analysis about the data
df.describe()
```

```python
import pandas as pd
import seaborn as sns

# Assuming df is your DataFrame
# Convert numerical columns to the appropriate data type
df['Sepal length'] = pd.to_numeric(df['Sepal length'], errors='coerce')
df['Sepal width'] = pd.to_numeric(df['Sepal width'], errors='coerce')
df['Petal length'] = pd.to_numeric(df['Petal length'], errors='coerce')
df['Petal width'] = pd.to_numeric(df['Petal width'], errors='coerce')

# Visualize the pairplot if the DataFrame contains numerical columns
sns.pairplot(df, hue='Class_labels')
```

```python
# Seperate features and target
data = df.values
X = data[:,0:4]
Y = data[:,4]
```

```python
import numpy as np

# Assuming X is your feature matrix and Y is your class labels
# Calculate average of each feature for all classes
Y_Data = np.array([np.average(X[:, i][Y == j].astype('float32')) for i in range(X.shape[1]) for j in np.unique(Y)])

# Reshape Y_Data into desired shape
Y_Data_reshaped = Y_Data.reshape(-1, len(np.unique(Y)))

# Swap axes if necessary
Y_Data_reshaped = np.swapaxes(Y_Data_reshaped, 0, 1)

# Assuming 'columns' is the list of feature names
X_axis = np.arange(len(columns) - 1)
width = 0.25  # Width for plotting

# Now you can use Y_Data_reshaped, X_axis, and width for plotting or further analysis

```

```python
# Plot the avarage
plt.bar(X_axis, Y_Data_reshaped[0], width, label = 'Setosa')
plt.bar(X_axis+width, Y_Data_reshaped[1], width, label = 'Versicolour')
plt.bar(X_axis+width*2, Y_Data_reshaped[2], width, label = 'Virginica')
plt.xticks(X_axis, columns[:4])
plt.xlabel("Features")
plt.ylabel("Value in cm.")
plt.legend(bbox_to_anchor=(1.3,1))
plt.show()
```

```python
# Split the data to train and test dataset.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.svm import SVC

# Assuming X_train and y_train are your training data
# Impute missing values in X_train
imputer = SimpleImputer(strategy='mean')  # You can choose other strategies as well
X_train_imputed = imputer.fit_transform(X_train)

# Instantiate and fit SVC model
svc = SVC()
svc.fit(X_train_imputed, y_train)
```
